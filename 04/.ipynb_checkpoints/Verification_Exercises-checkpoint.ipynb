```python
import pandas as pd 
import math
from datetime import datetime,date
import datetime
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
```

```python
st1_temp = pd.read_csv("https://raw.githubusercontent.com/maltemuellerm/GEO4902_2021/master/04/data/99735temp.csv")                 
st2_temp = pd.read_csv("https://raw.githubusercontent.com/maltemuellerm/GEO4902_2021/master/04/data/99754temp.csv")
st1_wind = pd.read_csv("https://raw.githubusercontent.com/maltemuellerm/GEO4902_2021/master/04/data/99735wind.csv")                 
st2_wind = pd.read_csv("https://raw.githubusercontent.com/maltemuellerm/GEO4902_2021/master/04/data/99754wind.csv") 
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

### Time series plots

```python
fig = plt.figure(figsize=(10,10))

ax1 = plt.subplot(511)
ax1.plot(st1_temp.time, st1_temp.obs, '-')
ax1.plot(st1_temp.time, st1_temp.model-271.5, '-')
#ax1.plot(wpoltime,era5.strd/3600,'-')
#ax1.set_xlim(tstart,tend)
#ax1.set_ylim(120,270)
#plt.title('Downwelling long-wave radiation')

ax1 = plt.subplot(512)
ax1.plot(st2_temp.time, st2_temp.obs, '-')
ax1.plot(st2_temp.time, st2_temp.model-271.5, '-')

ax1 = plt.subplot(513)
ax1.plot(st1_wind.time, st1_wind.obs, '-')
ax1.plot(st1_wind.time, st1_wind.model, '-')

ax1 = plt.subplot(514)
ax1.plot(st2_wind.time, st2_wind.obs, '-')
ax1.plot(st2_wind.time, st2_wind.model, '-')
```

### Plot observation vs error to find out on "conditional error"

```python
fig = plt.figure(figsize=(10,10))

ax1 = plt.subplot(221)
ax1.plot(st1_temp.obs, st1_temp.model-271.5-st1_temp.obs , '.')
ax1.plot(st1_temp.obs, st1_temp.model-271.5-st1_temp.obs , '.')
ax1.set_xlim(-6,6); ax1.set_ylim(-6,6)
plt.grid(True)

ax1 = plt.subplot(222)
ax1.plot(st2_temp.obs, st2_temp.model-271.5-st2_temp.obs , '.')
ax1.plot(st2_temp.obs, st2_temp.model-271.5-st2_temp.obs , '.')
ax1.set_xlim(-6,6); ax1.set_ylim(-6,6)
plt.grid(True)

ax1 = plt.subplot(223)
ax1.plot(st1_wind.obs, st1_wind.model-st1_wind.obs , '.')
ax1.plot(st1_wind.obs, st1_wind.model-st1_wind.obs , '.')
ax1.set_xlim(0,10); ax1.set_ylim(-10,10)
plt.grid(True)

ax1 = plt.subplot(224)
ax1.plot(st2_wind.obs, st2_wind.model-st2_wind.obs , '.')
ax1.plot(st2_wind.obs, st2_wind.model-st2_wind.obs , '.')
ax1.set_xlim(0,10); ax1.set_ylim(-10,10)
plt.grid(True)
```

### Bias, RMSE

```python
rmse1_t = np.sqrt(mean_squared_error(np.array(st1_temp.obs),np.array(st1_temp.model)-273.15))
rmse2_t = np.sqrt(mean_squared_error(np.array(st2_temp.obs),np.array(st2_temp.model)-273.15))
rmse1_w = np.sqrt(mean_squared_error(np.array(st1_wind.obs),np.array(st1_wind.model)))
rmse2_w = np.sqrt(mean_squared_error(np.array(st2_wind.obs),np.array(st2_wind.model)))

bias1_t  =np.mean(np.array(st1_temp.model)-273.15-np.array(st1_temp.obs))
bias2_t  =np.mean(np.array(st2_temp.model)-273.15-np.array(st2_temp.obs))
bias1_w  =np.mean(np.array(st1_wind.model)-np.array(st1_wind.obs))
bias2_w  =np.mean(np.array(st2_wind.model)-np.array(st2_wind.obs))

print ('        RMSE            BIAS')
print ('Temperature')
print ('#1 ',rmse1_t,bias1_t)
print ('#2 ',rmse2_t,bias2_t)
print ('Wind')
print ('#1 ',rmse1_w,bias1_w)
print ('#2 ',rmse2_w,bias2_w)
```

### RMSE without bias

```python
rmse1_t = np.sqrt(mean_squared_error(np.array(st1_temp.obs),np.array(st1_temp.model)-273.15-bias1_t))
rmse2_t = np.sqrt(mean_squared_error(np.array(st2_temp.obs),np.array(st2_temp.model)-273.15-bias2_t))
rmse1_w = np.sqrt(mean_squared_error(np.array(st1_wind.obs),np.array(st1_wind.model)-bias1_w))
rmse2_w = np.sqrt(mean_squared_error(np.array(st2_wind.obs),np.array(st2_wind.model)-bias2_w))
print ('        RMSE            BIAS')
print ('Temperature')
print ('#1 ',rmse1_t,bias1_t)
print ('#2 ',rmse2_t,bias2_t)
print ('Wind')
print ('#1 ',rmse1_w,bias1_w)
print ('#2 ',rmse2_w,bias2_w)
```

### Smooth the timeseries of the model

```python
from scipy.signal import savgol_filter
#result = savgol_filter(np.array(st2_wind.model), 25, 2)
st1_windsmooth = savgol_filter(np.array(st1_wind.model-bias1_w), 25, 1)
st2_windsmooth = savgol_filter(np.array(st2_wind.model-bias2_w), 25, 1)
```

```python
fig = plt.figure(figsize=(10,10))

ax1 = plt.subplot(511)
ax1.plot(st1_wind.time, st1_windsmooth, '-')
ax1.plot(st1_wind.time, st1_wind.model-bias1_w, '-')

ax1 = plt.subplot(512)
ax1.plot(st2_wind.time, st2_windsmooth, '-')
ax1.plot(st2_wind.time, st2_wind.model-bias2_w, '-')
```

```python
rmse1_w_smooth = np.sqrt(mean_squared_error(np.array(st1_wind.obs),np.array(st1_windsmooth)))
rmse2_w_smooth = np.sqrt(mean_squared_error(np.array(st2_wind.obs),np.array(st2_windsmooth)))
print ('RMSE smoothed vs. original')
print ("#1  ",rmse1_w_smooth,rmse1_w)
print ("#2  ",rmse2_w_smooth,rmse2_w)
```
